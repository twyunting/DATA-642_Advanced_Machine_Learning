```python
def func1(noise):
  x1 = random.uniform(1, 10) # The random variables x1 and x2 are over the interval [0, 10]
  x2 = random.uniform(1, 10)
  a = np.array([0.25, -0.25, 0.25], dtype=object) # theta0, theta1, and theta2
  b = np.matrix([[x1], [x2], [1]], dtype=object)
  ans = (np.dot(a, b)) + noise
  return ans

# if noise = 2
print(func1(2))
```
